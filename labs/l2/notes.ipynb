```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

X = np.array([
    [7,2], 
    [2,4], 
    [2, 3], 
    [3, 6], 
    [4, 4], 
    [9, 4], 
    [6, 8], 
    [9, 5], 
    [8, 7], 
    [10, 8]])
plt.figure(figsize=(8, 8))
for x in X:
    plt.scatter(x[0], x[1])
plt.show()
```

```python
# ZERO MEAN
X = X - np.mean(X)
X
```

```python
Sigma = np.cov(X, rowvar=False)
Sigma
```

```python
eigenvals, eigenvectors = np.linalg.eigh(Sigma)
eigenvals, eigenvectors
```

```python
max_eigen_vector = eigenvectors[:, np.argmax(eigenvals)]
max_eigen_vector
```

```python
Z = X * max_eigen_vector
Z
```

```python
plt.figure(figsize=(8, 8))
for z in Z:
    plt.scatter(z[0], z[1])
plt.show()
```
